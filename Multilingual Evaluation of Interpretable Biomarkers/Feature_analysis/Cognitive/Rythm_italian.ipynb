```python
from ultils import *
import os
import pandas as pd
import numpy as np
import seaborn as sns
from scipy.stats import kurtosis, skew
```

```python
# load csv files containing the first word time stamp.
word_analysis = []
path = os.listdir("/export/b15/afavaro/Frontiers/Italian_PD/Alignment/Whole_Align/")
for elem in path:
    if "PR1" not in elem:
        word_analysis.append(os.path.join('/export/b15/afavaro/Frontiers/Italian_PD/Alignment/Whole_Align/', elem))

list_of_dataframes = []
for element in word_analysis:
    participant = pd.read_csv(element)
    elem = os.path.basename(element)
    name = elem.split("_", -1)[-1]#.split(".csv")

    total = elem.split(".csv")[0]
    label = elem.split("_")[0]
    task = (elem.split("_", -1)[-1][:2])
    
    tot= participant.time_stamp.tolist()
    tot = (list(set(tot)))
    unique = participant['time_stamp'].unique()
    stdx = unique.std()
    kurtx = kurtosis(unique)
    skewx =  skew(unique)
    dict = {'name': total, 'label': label, 'tasks': task,'rhythm_std': stdx,'rhythm_skew': skewx, 'rhythm_kurt': kurtx } # 'Rhythm_kurt': kurtosis, 
    df = pd.DataFrame(dict, index=[0])
    list_of_dataframes.append(df)

data_ = pd.concat(list_of_dataframes, ignore_index=True)
data_ = data_.dropna()
```

```python
# group the data frame by task
# B1 and B2 are the two read passages.
# FB is the text dependent utterances.
task_grouped_= data_.groupby('tasks')
B1 = task_grouped_.get_group("B1")
B1.sort_values(by='name')
B2 = task_grouped_.get_group("B2")
B2.sort_values(by='name')
FB1 = task_grouped_.get_group("FB")
FB1.sort_values(by='name')
```

## First read passage

```python
grouped = B1.groupby("label") #here we extract the values by category
control = grouped.get_group("CN") 
parkinson_ = grouped.get_group("PD")
task = B1.columns[3:].values.tolist()
parkinson_all_feat = np.array([parkinson_[feat] for feat in np.array(B1.columns[3:])])
control_all_feat = np.array([control[feat] for feat in np.array(B1.columns[3:])])

with open('/export/b15/afavaro/Frontiers/submission/ITALIAN/COGNTIVE/Results/rhythm_b1_rt.txt', 'w') as f:
 
    kruskal(f, task, control_all_feat, parkinson_all_feat, "controls vs.", "parkinson")

kruskal_1 = read_stats_test('/export/b15/afavaro/Frontiers/submission/ITALIAN/COGNTIVE/Results/rhythm_b1_rt.txt')
holm_correction(kruskal_1)
```

```python
for i, title in enumerate(task):  
    
    nome = title
    controls = []
    parkinson=[]
    data = np.concatenate([control_all_feat[i], parkinson_all_feat[i]])
    data = data.tolist()
    data = [float(x) for x in data if x != "None"]
    [controls.append('CN') for value in range(len(control_all_feat[i]))]
    [parkinson.append('PD') for value in range(len(parkinson_all_feat[i]))]
    lista = controls+parkinson
    dict = {nome: data, 'Group': lista} 
    df = pd.DataFrame(dict)
    
    sns.set(font_scale=2)
    sns.catplot(x=nome, y="Group", kind="boxen", data=df, k_depth='full', palette="Blues",\
               height=4.2, aspect=1.4, showfliers = False)
```

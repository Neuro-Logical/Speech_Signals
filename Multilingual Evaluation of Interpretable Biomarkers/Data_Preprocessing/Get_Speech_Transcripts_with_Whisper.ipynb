```python
import os
import pandas as pd
import subprocess

pd.options.display.max_rows = 100
pd.options.display.max_colwidth = 1000
```

```python
''' Extract speech transcripts using Whisper: https://openai.com/blog/whisper/
1) For Italian use --language Italian
2) For Czech use --language Czech
3) For German use --language German
4) for Spanish use --language Spanish '''

path_recordings = '/export/b15/afavaro/Frontiers/NLS/NLS_RESAMPLED/'
path_recordings_all = [os.path.join(path_recordings, base) for base in os.listdir(path_recordings)]
for pa in path_recordings_all:
    print(pa)
    subprocess.run(f"whisper {pa} --language English", shell=True)
```
